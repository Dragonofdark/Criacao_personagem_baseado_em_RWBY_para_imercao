```python
import google.generativeai as genai
from google.colab import userdata
from nltk.tokenize import word_tokenize
from textblob import TextBlob

def criar_personagem():
  """
  Função para criar um personagem de RPG.

  Retorna:
    Um dicionário com as informações do personagem.
  """

  personagem = {
      "Nome": "",
      "Nível": 1,
      "Idade": 0,
      "Time": "",
      "EUO": 0,
      "P.O": 0,
      "Atributos": {
          "Força": 0,
          "Destreza": 0,
          "Agilidade": 0,
          "Percepção": 0,
          "Inteligência": 0,
          "Vigor": 0,
          "Espírito": 0,
          "Carisma": 0,
          "Manipulação": 0,
          "Sorte": 0,
          "Vida": 0,
      },
      "Habilidades": {
          "Combate Corporal": 0,
          "Combate à Distância": 0,
          "Esquiva": 0,
          "Ler Movimento": 0,
      },
      "Ações": {
          "Iniciativa": 0,
          "Reação de defesa": 0,
          "Deslocamento": 0,
      },
      "Pericias": {
          "Acrobacia": 0,
          "Lidar com animais": 0,
          "Arte": 0,
          "Mecanismo": 0,
          "Atletismo": 0,
          "Medicina": 0,
          "Ciências Naturais": 0,
          "Ocultismo": 0,
          "Concentração": 0,
          "Prestidigitação": 0,
          "Cultura": 0,
          "Procurar": 0,
          "Disfarces": 0,
          "Prontidão": 0,
          "Escapar": 0,
          "Rastrear": 0,
          "Furtividade": 0,
          "Venefício": 0,
      },
      "Pontos": 0,
      "Raça": "",
      "Semblence": [],
      "Estilo de Luta": "",
  }

  print("Bem-vindo à criação de personagem!")

  # Nome
  personagem["Nome"] = input("Digite o nome do seu personagem: ")

  # Idade
  while True:
    try:
      personagem["Idade"] = int(input("Digite a idade do seu personagem: "))
      break
    except ValueError:
      print("Por favor, digite um número inteiro válido para a idade.")

  # Time
  personagem["Time"] = input("Digite o time do seu personagem: ")

  # Atributos
  print("\nDistribua 10 pontos entre os atributos:")
  atributos = list(personagem["Atributos"].keys())
  pontos_atributos = 10
  while pontos_atributos > 0:
    print("\nPontos restantes:", pontos_atributos)
    for i, atributo in enumerate(atributos):
      print(f"{i+1}. {atributo}: {personagem['Atributos'][atributo]}")

    while True:
      try:
        escolha = int(input("Escolha o atributo para aumentar (1-11): ")) - 1
        if 0 <= escolha < len(atributos):
          break
        else:
          print("Opção inválida. Escolha um número de 1 a 11.")
      except ValueError:
        print("Por favor, digite um número inteiro válido.")

    while True:
      try:
        pontos = int(input(f"Quantos pontos deseja adicionar em {atributos[escolha]}? "))
        if 0 < pontos <= pontos_atributos:
          break
        else:
          print(f"Digite um valor entre 1 e {pontos_atributos}.")
      except ValueError:
        print("Por favor, digite um número inteiro válido.")

    personagem["Atributos"][atributos[escolha]] += pontos
    pontos_atributos -= pontos

  # Raça

  print("\nEscolha a Raça de seu Personagem: ")
  raca_disponiveis = ["Humano", "Fauni", "Gynoid (Android)", "Grim", "Dragão Primordial (Não Existe na obra original é criação minha)"]

  for i, raca in enumerate(raca_disponiveis):
    print(f"{i+1}. {raca}")

  while True:
      try:
          escolha = int(input("Digite o número da raça desejada: ")) - 1
          if 0 <= escolha < len(raca_disponiveis):
              sua_raca = raca_disponiveis[escolha]
              print(f"Sua raça é: {sua_raca}")
              break
          else:
              print("Opção inválida. Escolha um número da lista.")
      except ValueError:
          print("Por favor, digite um número inteiro válido.")

  # Semblence
  print("\nCrie seu semblece")
  personagem["Semblence"] = input("Seu Semblece: ")

  # Estilo de Luta
  print("\nInforme seu Estilo de Luta")
  personagem["Estilo de Luta"] = input("Seu estilo de Luta: ")

  # Habilidades
  print("\nDistribua 5 pontos entre as habilidades: ")
  habilidades = list(personagem["Habilidades"].keys())
  pontos_habilidades = 5
  while pontos_habilidades > 0:
    print("\nPontos restantes:", pontos_habilidades)
    for i, habilidade in enumerate(habilidades):
        print(f"{i+1}. {habilidade}: {personagem['Habilidades'][habilidade]}")

    while True:
        try:
            escolha = int(input("Escolha a habilidade para aumentar (1-4): ")) - 1
            if 0 <= escolha < len(habilidades):
                break
            else:
                print("Opção inválida. Escolha um número de 1 a 4.")
        except ValueError:
            print("Por favor, digite um número inteiro válido.")

    while True:
        try:
            pontos = int(input(f"Quantos pontos deseja adicionar em {habilidades[escolha]}? "))
            if 0 < pontos <= pontos_habilidades:
                break
            else:
                print(f"Digite um valor entre 1 e {pontos_habilidades}.")
        except ValueError:
            print("Por favor, digite um número inteiro válido.")

    personagem["Habilidades"][habilidades[escolha]] += pontos
    pontos_habilidades -= pontos

  # Ações
  print("\nDistribua 3 pontos entre as ações:")
  acoes = list(personagem["Ações"].keys())
  pontos_acoes = 3

  while pontos_acoes > 0:
    print("\nPontos restantes:", pontos_acoes)
    for i, acao in enumerate(acoes):
        print(f"{i+1}. {acao}: {personagem['Ações'][acao]}")

    while True:
        try:
            escolha = int(input("Escolha a ação para aumentar (1-3): ")) - 1
            if 0 <= escolha < len(acoes):
                break
            else:
                print("Opção inválida. Escolha um número de 1 a 3.")
        except ValueError:
            print("Por favor, digite um número inteiro válido.")

    while True:
        try:
            pontos = int(input(f"Quantos pontos deseja adicionar em {acoes[escolha]}? "))
            if 0 < pontos <= pontos_acoes:
                personagem["Ações"][acoes[escolha]] += pontos
                pontos_acoes -= pontos
                break # Adicione o 'break' aqui para sair do loop interno
            else:
                print(f"Digite um valor entre 1 e {pontos_acoes}.")
        except ValueError:
            print("Por favor, digite um número inteiro válido.")

        personagem["Ações"][acoes[escolha]] += pontos
        pontos_acoes -= pontos

  # Perícias
  print("\nEscolha 4 perícias para ter +2: ")
  pericias = list(personagem["Pericias"].keys())
  escolhas_pericias = 4
  while escolhas_pericias > 0:
    print("\nPerícias disponíveis:")
    for i, pericia in enumerate(pericias):
        print(f"{i+1}. {pericia}: {personagem['Pericias'][pericia]}")

    while True:
        try:
            escolha = int(input("Escolha uma perícia para aumentar (1-18): ")) - 1
            if 0 <= escolha < len(pericias):
                break
            else:
                print("Opção inválida. Escolha um número de 1 a 18.")
        except ValueError:
            print("Por favor, digite um número inteiro válido.")

    personagem["Pericias"][pericias[escolha]] += 2
    escolhas_pericias -= 1

  print("\nPersonagem criado com sucesso!")
  return personagem

def chatbot(personagem):
    """
    Função para interagir com o chatbot.

    Args:
        personagem: Um dicionário com as informações do personagem.
    """
    print(f"\nOlá, {personagem['Nome']}! O que você deseja fazer?")
    while True:
        comando = input("> ")
        if comando.lower() in ["sair", "quit", "exit", "fim"]:
            print("Saindo da aplicação. Até mais!")
            break

        # Analisar o comando do usuário
        tokens = word_tokenize(comando)  # Define 'tokens' aqui

        # Realizar ações com base na resposta
        if "atacar" in tokens:
            # Simulação de ataque
            dano = personagem["Atributos"]["Força"] + personagem["Habilidades"]["Combate Corporal"]
            print(f"{personagem['Nome']} ataca! Causa {dano} de dano.")
            descricao_ataque = chat.send_message("Descreva em portugues do Brasil um monstro sendo derrotado", generation_config=generation_conf, safety_settings=conf_seguranca)
            print(descricao_ataque.text)
        elif "defender" in tokens:
            # Simulação de uma defesa
            print(f"{personagem['Nome']} se defende!")
            descricao_defesa = chat.send_message(f"Descreva em portugeus do Brasil o {personagem['Nome']} se defendendo de o ataque de um monstro.", generation_config=generation_conf, safety_settings=conf_seguranca)
            print(descricao_defesa.text)
        elif "explorar" in tokens:
            # Simulando uma exploração em uma area
            print(f"{personagem['Nome']} explorou a área...")
            descricao_area = chat.send_message("Descreva em portugeus do Brasil uma floresta densa e misteriosa.", generation_config=generation_conf, safety_settings=conf_seguranca)
            print(descricao_area.text)
        elif "falar" in tokens:
            # Simulando com uma conversa
            print(f"{personagem['Nome']} tenta iniciar uma conversa.")
            descricao_conversa = chat.send_message(f"Descreva em portugeus do Brasil o {personagem['Nome']} conversando com sua parceira de equipe Jack.", generation_config=generation_conf, safety_settings=conf_seguranca)
            print(descricao_conversa.text)
        elif "usar" in tokens:
            # Simulando o uso de um item
            print(f"{personagem['Nome']} tenta usar um item.")
            descricao_item = chat.send_message(f"Descreva em portugeus do Brasil o {personagem['Nome']} utilizando uma pedra chamada dust para destruir um obstaculo.", generation_config=generation_conf, safety_settings=conf_seguranca)
            print(descricao_item.text)
        else:
            # Processar comandos usando o modelo de linguagem (apenas se não for uma ação predefinida)
            resposta = chat.send_message(comando, generation_config=generation_conf, safety_settings=conf_seguranca)
            print(f"\n{resposta.text}") # Exibe a resposta do modelo



# Criar personagem
personagem = criar_personagem()

# Iniciar chatbot
chatbot(personagem)
```
